```python
import sklearn
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv("C:/Users/LocalAdminRed/Desktop/data_r_lineal.csv", sep=",")
```

```python
data
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(data["metro"],data["precio"])
```

```python
data.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
regresion = linear_model.LinearRegression()
```

```python
metro = data["metro"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(metro, data["precio"])
print("interseccion (b)", modelo.intercept_)
print("pendiente (m)", modelo.coef_)
```

```python
entrada =[[5],[15],[20],[25]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
data.plot.scatter(x="metro", y="precio", label ='datos')
plt.scatter(entrada, predicciones, color='green')
plt.plot(entrada, predicciones, color='black', label='linea de regresion')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
